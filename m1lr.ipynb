```python
X=df.drop("income",axis=1)
y=df.income.values
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
```

```python
from sklearn.preprocessing import LabelEncoder
cat=[i  for i in X.columns if df[i].dtypes == "O"]
```

```python
cat
```

```python
model1=LabelEncoder()
for i in cat:
    X_train[i]=model1.fit_transform(X_train[i])
    X_test[i]=model1.fit_transform(X_test[i])
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
from sklearn.preprocessing import StandardScaler
model2=StandardScaler()
X_train=model2.fit_transform(X_train)
X_test=model2.fit_transform(X_test)
```

```python
X_test[0:5]
```

```python
X_train=pd.DataFrame(X_train,columns=X.columns)
```

```python
X_test=X_test.values
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

main_model=LogisticRegression()
main_model.fit(X_train.values,y_train)
```

```python
y_pred=main_model.predict(X_test)
```

```python
y_pred
```

```python
print("logistic regression acuracy score with all features {0:0.4f}".format(accuracy_score(y_pred,y_test)))
```
